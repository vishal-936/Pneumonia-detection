```python
from keras.models import load_model
from tensorflow.keras.utils import load_img, img_to_array 
import numpy as np
import cv2
```

```python
model = load_model('/content/gdrive/MyDrive/Colab Notebooks/Pneumonia Prediction/chest_xray/Pneumonia_detection_model.h5')
```

```python
def predict(img_path):
  img = load_img(img_path,target_size = (150,150),color_mode = 'grayscale')
  img = img_to_array(img)
  img *= 1./255
  img = np.expand_dims(img,axis = 0)
  return model.predict(img)[0][0]
```

```python
predict('/content/gdrive/MyDrive/Colab Notebooks/Pneumonia Prediction/chest_xray/test/NORMAL/IM-0005-0001.jpeg')
```
